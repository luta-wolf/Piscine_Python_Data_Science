```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
try:
    connection = sqlite3.connect('../data/checking-logs.sqlite')
    print("Successfully connect to the database")
except sqlite3.Error as error:
    print("Error when connecting to the database", error)
```

```python
query = "select uid, timestamp from checker where uid like 'user_%'"
df = pd.io.sql.read_sql(query, connection, parse_dates=['timestamp'])
connection.close()
df
```

```python
df['hour'] = df['timestamp'].dt.hour
df['weekday'] = df['timestamp'].apply(lambda x: x.strftime('%a'))
df = df.drop('timestamp', axis=1)
df
```

```python
columns = df.groupby('uid').count().sort_values('hour', ascending=False).index
```

```python
table_1 = df.groupby('uid')['weekday'].value_counts().unstack(0)
table_1 = table_1.reindex(['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
table_1 = table_1.reindex(columns=columns).fillna(0)
table_1
```

```python
table_2 = df.groupby('uid').hour.value_counts().unstack(0)
table_2 = table_2.reindex(index=range(24), fill_value=0)
table_2 = table_2.reindex(columns=columns).rename_axis('hour').fillna(0)
table_2
```

```python
fig1, ax1 = plt.subplots(figsize=(15, 8))
im1 = ax1.imshow(table_1, cmap='RdBu')
cax1 = make_axes_locatable(ax1).append_axes('right', size='5%', pad=0.1)
plt.colorbar(im1, cax=cax1)
ax1.set_xticks(range(table_1.shape[1]))
ax1.set_xticklabels(table_1.columns)
ax1.set_yticks(range(table_1.shape[0]))
ax1.set_yticklabels(table_1.index)
plt.setp(ax1.get_xticklabels(), rotation=90)

fig2, ax2 = plt.subplots(figsize=(15, 10))
im2 = ax2.imshow(table_2, cmap='jet')
cax2 = make_axes_locatable(ax2).append_axes('right', size='5%', pad=0.1)
plt.colorbar(im2, cax=cax2)
ax2.set_xticks(range(table_2.shape[1]))
ax2.set_xticklabels(table_2.columns)
ax2.set_yticks(range(table_2.shape[0]))
ax2.set_yticklabels(table_2.index)
plt.setp(ax2.get_xticklabels(), rotation=90)

plt.show()
```

### ◦ “Which user has the most commits on Tue?”
The answer: user_2.
### ◦ “Which user has the most commits on Thu?”
The answer: user_4.
### ◦ “On which weekday do the users not like making a lot of commits?”
The answer: Fri.
### ◦ “Which user at which hour made the largest number of commits?”
The answer: user_4, 21.
