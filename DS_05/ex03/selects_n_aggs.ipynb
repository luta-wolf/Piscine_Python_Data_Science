```python
import pandas as pd
```

• load the JSON file that you created in the previous exercise into a dataframe<br>
◦ set CarNumber as the index column

```python
df = pd.read_json('../ex02/auto.json')
df.set_index('CarNumber', inplace=True)
df
```

◦ display the rows only where the fines are more than 2,100

```python
df[df['Fines'] > 2100]
```

◦ display the rows only where the fines are more than 2,100 and the refund equals 2

```python
df[(df['Fines'] > 2100) & (df['Refund'] == 2)]
```

◦ display the rows only where the models are from the list: [’Focus’, ’Corolla’]

```python
df[(df['Model'] == 'Focus') | (df['Model'] == 'Corolla')]
```

◦ display the rows only where the car number is from the list: [’Y7689C197RUS’, ’92928M178RUS’, ’7788KT197RUS’, ’H115YO163RUS’, ’X758HY197RUS’]

```python
df[df.index.isin(['Y7689C197RUS', '92928M178RUS', '7788KT197RUS', 'H115YO163RUS', 'X758HY197RUS'])]
```

• make the aggregations with the make and the model<br>
◦ display the median fines grouped by the make

```python
df.groupby('Make')['Fines'].median()
```

◦ display the median fines grouped by the make and the model

```python
df.groupby(['Make', 'Model'])['Fines'].median()
```

◦ display the number of fines grouped by the make and the model in order to
understand if we can trust the median values

```python
df.groupby(['Make', 'Model'])['Fines'].count()
```

◦ display the minimum and the maximum fines grouped by the make and the
model in order to better understand the variance

```python
min = df.groupby(['Make', 'Model'])['Fines'].min()
max = df.groupby(['Make', 'Model'])['Fines'].max()
min, max
```

◦ display the standard deviation of the fines grouped by the make and the model
in order to better understand the variance

```python
df.groupby(['Make', 'Model'])['Fines'].std()
```

• make the aggregations with the car number<br>
◦ display the car numbers grouped by the number of the fines in descending
order, we want to find those who most often violated the law

```python
df.groupby(df.index)['Fines'].count().sort_values(ascending=False)
```

◦ select from the initial dataframe all the rows corresponding to the top-1 car
number, we want to zoom in a little bit

```python
df[df.index == df.index[0]]
```

◦ display the car numbers grouped by the sum of the fines in descending order,
we want to find those who paid the most

```python
ds = df.groupby(df.index)['Fines'].sum().sort_values(ascending=False)
ds
```

◦ select from the initial dataframe all the rows corresponding to the top-1 car
number, we want to zoom in a little bit

```python
df[df.index == ds.index[0]]
```

◦ display a table that answers the question: are there any car numbers that were
connected to different models?

```python
ds = df.reset_index()
ds = ds.drop_duplicates(['CarNumber', 'Model'])
ds = ds.groupby('CarNumber')['Make'].count().sort_values(ascending=False)
df[df.index == ds.index[0]]
```
