• count how many rows satisfy the following conditions using only one query with
any number of subqueries:<br>
◦ count the rows from the pageviews table but only with users from the checker
table with:<br>
∗ status = ’ready’, we do not want to analyze the logs that are in status
checking<br>
∗ numTrials = 1, we want to analyze only the first commits, because only
they can tell us when a student started working on a lab<br>
∗ labnames should be from the list: ’laba04’, ’laba04s’, ’laba05’, ’laba06’,
’laba06s’, ’project1’. Only they were active during the experiment<br>
◦ store in the dataframe checkers with the column cnt

```python
query = "SELECT COUNT(*) as cnt FROM \
(SELECT uid FROM pageviews \
WHERE uid IN \
(SELECT uid FROM checker \
WHERE status='ready' AND numtrials=1 \
AND labname IN ('laba04', 'laba04s', 'laba05', \
'laba06', 'laba06s', 'project1')))"
checkers = pd.io.sql.read_sql(query, db_conn)
checkers
```
